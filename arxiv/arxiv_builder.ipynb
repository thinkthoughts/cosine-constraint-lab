```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def paper_hydration(problem=1.0, hypothesis=0.9, methods=0.7, results=0.9,
                    alternatives=0.3, limitations=0.5, predictions=0.8, testability=0.6):
    """Calculate paper hydration based on section scores."""
    scores = np.array([problem, hypothesis, methods, results, 
                       alternatives, limitations, predictions, testability])
    basis = np.ones(8)
    cos = np.dot(scores, basis) / (np.linalg.norm(scores) * np.linalg.norm(basis))
    angle = np.arccos(np.clip(cos, -1.0, 1.0)) * 180 / np.pi
    hydration = 100 * (1 - angle / 90)
    
    if angle < 10:
        verdict = "✅ VC / GOS"
        color = "#2ecc71"
    elif angle < 35:
        verdict = "⚠️ weak VC"
        color = "#f1c40f"
    else:
        verdict = "❌ IA / ZOS"
        color = "#e74c3c"
    
    return angle, hydration, verdict, color

def update_paper(problem, hypothesis, methods, results, alternatives, limitations, predictions, testability):
    angle, hydration, verdict, color = paper_hydration(
        problem, hypothesis, methods, results, alternatives, limitations, predictions, testability
    )
    
    print("=" * 60)
    print("PAPER HYDRATION CALCULATOR")
    print("=" * 60)
    print(f"\n📐 Angle: {angle:.1f}°")
    print(f"💧 Hydration: {hydration:.1f}%")
    print(f"🏷️  Verdict: {verdict}")
    
    # Gauge
    fig, ax = plt.subplots(figsize=(10, 2))
    ax.barh([0], [angle], color=color, height=0.3, edgecolor='black')
    ax.axvline(10, color='#2ecc71', linestyle='--', alpha=0.7, label='VC Lock (10°)')
    ax.axvline(35, color='#f1c40f', linestyle=':', alpha=0.7, label='Weak VC (35°)')
    ax.axvline(45, color='#e74c3c', linestyle='--', alpha=0.7, label='Critical (45°)')
    ax.set_xlim(0, 90)
    ax.set_yticks([])
    ax.set_xlabel('Angle (°)')
    ax.set_title(f'Paper Hydration: {hydration:.0f}% — {verdict}')
    ax.legend(loc='upper right', fontsize=8)
    ax.grid(True, alpha=0.3, axis='x')
    plt.tight_layout()
    plt.show()

print("=" * 60)
print("PAPER BUILDER: S-star Formation via Black Hole Merger")
print("=" * 60)
print("\nAdjust the sliders to see how each section affects the paper's hydration.\n")
print("Target: Angle < 10°, Hydration > 90% (VC/GOS)\n")

interact(update_paper,
         problem=FloatSlider(min=0, max=1, step=0.05, value=1.0, description='Problem statement'),
         hypothesis=FloatSlider(min=0, max=1, step=0.05, value=0.9, description='Hypothesis'),
         methods=FloatSlider(min=0, max=1, step=0.05, value=0.7, description='Methods detail'),
         results=FloatSlider(min=0, max=1, step=0.05, value=0.9, description='Results'),
         alternatives=FloatSlider(min=0, max=1, step=0.05, value=0.3, description='Alternative explanations'),
         limitations=FloatSlider(min=0, max=1, step=0.05, value=0.5, description='Limitations'),
         predictions=FloatSlider(min=0, max=1, step=0.05, value=0.8, description='Predictions'),
         testability=FloatSlider(min=0, max=1, step=0.05, value=0.6, description='Testability'))
```
