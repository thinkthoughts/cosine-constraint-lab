## Core Claim

> *"A recent massive black hole merger with Sgr A* (2 × 10⁵ M⊙ within the last 10 Myr) can self-consistently produce many of the orbital properties of the Galactic nuclear star cluster within 0.5 pc. Stars near the inner edge of an eccentric disk migrate inward and are driven to high eccentricities and inclinations due to secular torques similar to the eccentric Kozai-Lidov mechanism."*

## Background Concepts

| Concept | Description |
|---------|-------------|
| **S-stars** | Young stars within 0.04 pc of Sgr A* (Galactic Center black hole) on eccentric, isotropically distributed orbits |
| **Gravitational wave recoil kick** | Asymmetric emission of gravitational waves during black hole merger imparts a kick to the remnant |
| **Eccentric Kozai-Lidov (EKL) mechanism** | Secular gravitational perturbations that drive oscillations in eccentricity and inclination |
| **N-body simulations** | Numerical modeling of gravitational dynamics of many particles |
| **Apse-aligned eccentric disk** | Disk where pericenters are aligned |

## Key Results

| Result | Significance |
|--------|--------------|
| Black hole merger produces apse-aligned eccentric disk | Explains observed disk structure within 0.5 pc |
| Stars at inner edge migrate inward via EKL-like torques | Populates the initially empty region <0.04 pc |
| Produces high eccentricity, high inclination S-stars | Matches observed orbital properties |
| Timescale: a few Myr | Consistent with young age of S-stars |
| Merger mass: 2^{+3}_{-1.2} × 10⁵ M⊙ | Specific, testable prediction |

````python
import numpy as np
import matplotlib.pyplot as plt

# Core functions
def angle_degrees(a, b):
    cos = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    return np.arccos(np.clip(cos, -1.0, 1.0)) * 180 / np.pi

def lock_status(angle):
    if angle < 10:
        return {"status": "VC / GOS", "color": "#2ecc71", "emoji": "🔒✅", "recommendation": "Well-constructed paper"}
    elif angle < 35:
        return {"status": "weak VC", "color": "#f1c40f", "emoji": "⚠️", "recommendation": "Needs stronger evidence"}
    elif angle <= 55:
        return {"status": "IA / ZOS", "color": "#e74c3c", "emoji": "🔴💀", "recommendation": "Vague claims, over-hyped"}
    else:
        return {"status": "broken", "color": "#95a5a6", "emoji": "⚰️", "recommendation": "Fundamentally flawed"}

def hydration_score(angle):
    score = max(0, min(100, 100 * (1 - angle / 90)))
    status = lock_status(angle)
    return {
        'hydration_percent': round(score, 1),
        'status': status['status'],
        'recommendation': status['recommendation'],
        'emoji': status['emoji']
    }

print("=" * 60)
print("PAPER ANALYSIS: S-star Formation via Black Hole Merger")
print("Authors: Akiba, Madigan, et al.")
print("arXiv:2410.19881 (2024)")
print("=" * 60)
````

````python
# Claim vector (8 dimensions)
claim_vector = np.array([1.0, 1.0, 1.0, 1.0, 0.95, 0.95, 0.9, 1.0])
constraint_basis = np.ones(8)

angle = angle_degrees(claim_vector, constraint_basis)
status = lock_status(angle)
hydration = hydration_score(angle)

constraint_names = [
    "C1: Problem clearly stated",
    "C2: Specific hypothesis",
    "C3: Methods specified",
    "C4: Mechanism identified",
    "C5: Quantitative prediction",
    "C6: Novelty",
    "C7: Testability",
    "C8: Anchoring to prior work"
]

print("\nConstraint adherence:\n")
for name, score in zip(constraint_names, claim_vector):
    bar = "█" * int(score * 20) + "░" * (20 - int(score * 20))
    print(f"  {name[:25]:25} {bar} {score:.0%}")

print(f"\n📐 Angle: {angle:.1f}°")
print(f"💧 Hydration: {hydration['hydration_percent']:.0f}%")
print(f"🏷️  Status: {status['status']} {status['emoji']}")
print(f"📋 Recommendation: {status['recommendation']}")
print(f"\n💡 Well-hydrated paper — clear problem, specific hypothesis, N-body simulations, identifies EKL mechanism, makes quantitative testable predictions.")
````

````python
fig, ax = plt.subplots(figsize=(10, 4))

angles_plot = np.linspace(0, 90, 100)
stability = np.cos(np.radians(angles_plot))

ax.plot(angles_plot, stability, 'k-', linewidth=2)
ax.axvline(angle, color=status['color'], linestyle='--', linewidth=2, label=f'Paper: {angle:.1f}°')
ax.axvline(10, color='#2ecc71', linestyle=':', alpha=0.7, label='VC Lock (10°)')
ax.axvline(45, color='#e74c3c', linestyle='--', alpha=0.7, label='Critical (45°)')
ax.fill_between(angles_plot, 0, stability, where=(angles_plot < 10), alpha=0.2, color='#2ecc71')
ax.fill_between(angles_plot, 0, stability, where=(angles_plot > 35) & (angles_plot < 55), alpha=0.15, color='#e74c3c')

ax.set_xlim(0, 90)
ax.set_ylim(0, 1)
ax.set_xlabel('Angle from Constraint Basis (degrees)')
ax.set_ylabel('cos(θ) — stability')
ax.set_title('Hydration Analysis: S-star Formation via Black Hole Merger')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
