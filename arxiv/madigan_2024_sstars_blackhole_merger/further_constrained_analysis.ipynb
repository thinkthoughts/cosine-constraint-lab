**Further-constrained angle:** ~8-12° → **weak VC / borderline**

```python
import numpy as np
import matplotlib.pyplot as plt

# Core functions
def angle_degrees(a, b):
    cos = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    return np.arccos(np.clip(cos, -1.0, 1.0)) * 180 / np.pi

def lock_status(angle):
    if angle < 10:
        return {"status": "VC / GOS", "color": "#2ecc71", "emoji": "🔒✅", "recommendation": "Well-constructed"}
    elif angle < 35:
        return {"status": "weak VC", "color": "#f1c40f", "emoji": "⚠️", "recommendation": "Needs strengthening"}
    elif angle <= 55:
        return {"status": "IA / ZOS", "color": "#e74c3c", "emoji": "🔴💀", "recommendation": "Vague claims"}
    else:
        return {"status": "broken", "color": "#95a5a6", "emoji": "⚰️", "recommendation": "Fundamentally flawed"}

def hydration_score(angle):
    score = max(0, min(100, 100 * (1 - angle / 90)))
    status = lock_status(angle)
    return {
        'hydration_percent': round(score, 1),
        'status': status['status'],
        'recommendation': status['recommendation'],
        'emoji': status['emoji']
    }

print("=" * 60)
print("FURTHER-CONSTRAINED ANALYSIS")
print("S-star Formation via Black Hole Merger")
print("=" * 60)
```

```python
# Stricter claim vector (9 dimensions)
# Lower scores reflect stricter standards
claim_vector = np.array([1.0, 0.9, 0.85, 0.9, 0.7, 0.9, 0.6, 0.5, 0.7])
constraint_basis = np.ones(9)

angle = angle_degrees(claim_vector, constraint_basis)
status = lock_status(angle)
hydration = hydration_score(angle)

constraint_names = [
    "C1: Problem clearly stated",
    "C2: Specific hypothesis",
    "C3: Methods specified",
    "C4: Mechanism identified",
    "C5: Quantitative prediction",
    "C6: Novelty",
    "C7: Testability",
    "C8: Alternative explanations",
    "C9: Observational constraints"
]

print("\nStricter Constraint Adherence:\n")
for name, score in zip(constraint_names, claim_vector):
    bar = "█" * int(score * 20) + "░" * (20 - int(score * 20))
    print(f"  {name[:27]:27} {bar} {score:.0%}")

print(f"\n📐 Further-constrained Angle: {angle:.1f}°")
print(f"💧 Hydration: {hydration['hydration_percent']:.0f}%")
print(f"🏷️  Status: {status['status']} {status['emoji']}")
print(f"📋 Recommendation: {status['recommendation']}")
print(f"\n💡 Under stricter constraints, the paper drops from VC to weak VC.")
print("   Main gaps: broad uncertainty range, distant testability, no alternative model comparison.")
```

```python
# Original scores for the 9 dimensions (estimated for C8, C9)
original_scores = [1.0, 1.0, 1.0, 1.0, 0.95, 0.95, 0.9, 0.6, 0.8]
stricter_scores = [1.0, 0.9, 0.85, 0.9, 0.7, 0.9, 0.6, 0.5, 0.7]
categories = constraint_names

x = np.arange(len(categories))
width = 0.35

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Bar chart comparison
ax1.bar(x - width/2, original_scores, width, label='Original (VC)', color='#2ecc71', alpha=0.7)
ax1.bar(x + width/2, stricter_scores, width, label='Further-Constrained', color='#f1c40f', alpha=0.7)
ax1.set_xticks(x)
ax1.set_xticklabels(categories, rotation=45, ha='right', fontsize=8)
ax1.set_ylim(0, 1)
ax1.set_ylabel('Constraint Score')
ax1.set_title('Constraint Score Comparison')
ax1.legend()
ax1.grid(True, alpha=0.3, axis='y')

# Plot 2: Gap analysis (where did the score drop most?)
gaps = [original_scores[i] - stricter_scores[i] for i in range(len(categories))]
colors_gap = ['#e74c3c' if g > 0.25 else '#f1c40f' if g > 0.1 else '#2ecc71' for g in gaps]

ax2.barh(categories, gaps, color=colors_gap, edgecolor='black')
ax2.set_xlim(0, 0.5)
ax2.set_xlabel('Score Gap (Original - Further-Constrained)')
ax2.set_title('Largest Gaps (Opportunities for Strengthening)')
ax2.axvline(0.1, color='#f1c40f', linestyle='--', alpha=0.7, label='Moderate gap threshold')
ax2.axvline(0.25, color='#e74c3c', linestyle='--', alpha=0.7, label='Large gap threshold')
ax2.legend(loc='lower right', fontsize=8)
ax2.grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.show()

print("\n" + "=" * 60)
print("GAP ANALYSIS")
print("=" * 60)
print("""
Largest gaps (further-constrained vs original):

1. C7 (Testability): -0.3 — LISA is far future; no near-term test
2. C5 (Quantitative prediction): -0.25 — Uncertainty range is broad
3. C3 (Methods specified): -0.15 — Simulation parameters could be more detailed
4. C8 (Alternative explanations): -0.1 — No discussion of competing models
""")
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

angles_plot = np.linspace(0, 90, 100)
stability = np.cos(np.radians(angles_plot))

ax.plot(angles_plot, stability, 'k-', linewidth=2)
ax.axvline(angle, color=status['color'], linestyle='--', linewidth=2, label=f'Further-Constrained: {angle:.1f}°')
ax.axvline(5, color='#2ecc71', linestyle=':', alpha=0.5, label='Original: ~5°')
ax.axvline(10, color='#2ecc71', linestyle='--', alpha=0.7, label='VC Lock (10°)')
ax.axvline(45, color='#e74c3c', linestyle='--', alpha=0.7, label='Critical (45°)')
ax.fill_between(angles_plot, 0, stability, where=(angles_plot < 10), alpha=0.2, color='#2ecc71')
ax.fill_between(angles_plot, 0, stability, where=(angles_plot > 35) & (angles_plot < 55), alpha=0.15, color='#e74c3c')

ax.set_xlim(0, 90)
ax.set_ylim(0, 1)
ax.set_xlabel('Angle from Constraint Basis (degrees)')
ax.set_ylabel('cos(θ) — stability')
ax.set_title('Hydration Analysis: Further-Constrained vs Original')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
