## 1. Setup

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from typing import Dict

# Colab setup
if 'google.colab' in sys.modules:
    !pip install numpy matplotlib ipywidgets > /dev/null 2>&1
    from IPython.display import display, HTML, clear_output
    print("✓ Colab environment ready")
else:
    print("✓ Local environment ready")

# Core functions
def angle_degrees(a: np.ndarray, b: np.ndarray) -> float:
    cos = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    return np.arccos(np.clip(cos, -1.0, 1.0)) * 180 / np.pi

def lock_status(angle: float) -> Dict[str, object]:
    if angle < 10:
        return {"status": "VC / GOS", "phase": "LOCKED", 
                "interpretation": "Valid Construction — well-hydrated claim",
                "color": "#2ecc71", "emoji": "🔒✅", "recommendation": "Accept"}
    elif angle < 35:
        return {"status": "weak VC", "phase": "DRIFTING", 
                "interpretation": "Partially constructed — needs more anchoring",
                "color": "#f1c40f", "emoji": "⚠️", "recommendation": "Revise"}
    elif angle <= 55:
        return {"status": "IA / ZOS", "phase": "CRITICAL", 
                "interpretation": "Invalid Assignment — collapses without construction",
                "color": "#e74c3c", "emoji": "🔴💀", "recommendation": "Reject"}
    else:
        return {"status": "weak IA / broken", "phase": "DECOUPLED", 
                "interpretation": "Unrecognizable claim — no constraint respect",
                "color": "#95a5a6", "emoji": "⚰️", "recommendation": "Discard"}

def hydration_score(angle: float) -> Dict:
    score = max(0, min(100, 100 * (1 - angle / 90)))
    status = lock_status(angle)
    return {
        'hydration_percent': round(score, 1),
        'status': status['status'],
        'recommendation': status['recommendation'],
        'emoji': status['emoji']
    }

print("✓ Setup complete")
print("\n" + "=" * 60)
print("MATH CLAIMS HYDRATION: PH & Omega-Categorical CSPs")
print("=" * 60)
```

## 6. Visualization: Three Claims on the Spectrum

```python
fig, ax = plt.subplots(figsize=(12, 6))

angles_plot = np.linspace(0, 90, 100)
stability = np.cos(np.radians(angles_plot))

ax.plot(angles_plot, stability, 'k-', linewidth=2.5, label='cos(θ) — stability')

# Shade zones
ax.axvspan(0, 10, alpha=0.2, color='#2ecc71', label='VC Lock Zone (<10°)')
ax.axvspan(35, 55, alpha=0.15, color='#e74c3c', label='Critical Zone (≈45°)')

# Mark claims
claims = [
    ("Your Seminar Claim (VC)", angle, '#2ecc71'),
    ("Bodirsky-Grohe (2008)", angle_bg, '#f1c40f'),
    ("Over-Extrapolation (IA)", angle_over, '#e74c3c')
]

for i, (label, ang, color) in enumerate(claims):
    cos_val = np.cos(np.radians(ang))
    ax.scatter(ang, cos_val, s=200, c=color, marker='D', edgecolors='black', zorder=5)
    ax.annotate(label, xy=(ang, cos_val), xytext=(ang + 3, cos_val - 0.08),
                fontsize=9, ha='left', bbox=dict(boxstyle='round', facecolor=color, alpha=0.2))

ax.axvline(10, color='#2ecc71', linestyle='--', alpha=0.7)
ax.axvline(45, color='#e74c3c', linestyle='--', linewidth=2, alpha=0.8, label='45° Critical Boundary')

ax.set_xlim(0, 90)
ax.set_ylim(0, 1)
ax.set_xlabel('Angle from Constraint Basis (degrees)', fontsize=12)
ax.set_ylabel('cos(θ) — stability', fontsize=12)
ax.set_title('Hydration Analysis: PH & Omega-Categorical CSP Claims', fontsize=14, fontweight='bold')
ax.legend(loc='upper right')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
