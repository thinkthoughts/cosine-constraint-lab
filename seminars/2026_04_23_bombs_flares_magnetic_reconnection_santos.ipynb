## Abstract

Magnetic reconnection is a fundamental plasma process that governs energy release in the solar atmosphere and throughout space plasmas. This talk emphasizes reconnection across solar scales, from Ellerman bombs to X-class flares, and explores whether a common mechanism operates across regimes.

Recent advances include plasmoid instability, turbulence, and results from DKIST observations and r-MHD simulations, with future DKIST + ALMA work constraining magnetic evolution and heating.

## Speaker's claim

Magnetic reconnection governs energy release across solar scales, with shared mechanisms suggested across regimes and constrained by observations and simulations.
