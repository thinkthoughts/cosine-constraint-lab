## Core Claim

> *"Condensed matter physics aims to explore quantum phenomena emerging from interactions between nuclei and electrons. Conventional crystals are often very complicated, making it hard to extract essential ingredients. Rhombohedral graphene — part of natural graphite — has the simplest chemistry and structure, yet can be controlled by experimental knobs to exhibit chiral superconductivity and the fractional quantum anomalous Hall effect. This establishes a new paradigm for studying emergent quantum phenomena."*

## Two Major Results

### 1. Chiral Superconductivity (Nature 2025)

| Property | Observation |
|----------|-------------|
| **Material** | Rhombohedral tetralayer and pentalayer graphene |
| **No moiré** | Pure carbon, no superlattice effects |
| **Tc** | Up to 300 mK |
| **Charge density** | As low as 2.4 × 10¹¹ cm⁻² |
| **Key signature** | Magnetic hysteresis in Rxx (TRSB) — absent in all other superconductors |
| **Critical field** | B⊥ = 1.4 T (highest for graphene superconductivity) |
| **Regime** | Strong-coupling, BCS-BEC crossover |

### 2. Fractional Quantum Anomalous Hall Effect (Nature 2024)

| Property | Observation |
|----------|-------------|
| **Material** | Rhombohedral pentalayer graphene–hBN moiré superlattice |
| **Zero magnetic field** | FQAHE without external B |
| **Fractional states** | v = 2/3, 3/5, 4/7, 4/9, 3/7, 2/5 |
| **Composite Fermi liquid** | Observed at v = 1/2 |
| **Implications** | Non-Abelian anyons, topological quantum computation |

````python
import numpy as np
import matplotlib.pyplot as plt

# Core functions
def angle_degrees(a, b):
    cos = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    return np.arccos(np.clip(cos, -1.0, 1.0)) * 180 / np.pi

def lock_status(angle):
    if angle < 10:
        return {"status": "VC / GOS", "color": "#2ecc71", "emoji": "🔒✅", "recommendation": "Accept"}
    elif angle < 35:
        return {"status": "weak VC", "color": "#f1c40f", "emoji": "⚠️", "recommendation": "Revise"}
    elif angle <= 55:
        return {"status": "IA / ZOS", "color": "#e74c3c", "emoji": "🔴💀", "recommendation": "Reject"}
    else:
        return {"status": "broken", "color": "#95a5a6", "emoji": "⚰️", "recommendation": "Discard"}

def hydration_score(angle):
    score = max(0, min(100, 100 * (1 - angle / 90)))
    status = lock_status(angle)
    return {
        'hydration_percent': round(score, 1),
        'status': status['status'],
        'recommendation': status['recommendation'],
        'emoji': status['emoji']
    }

print("=" * 60)
print("SEMINAR ANALYSIS: Emergent Quantum Phenomena in Graphene")
print("Speaker: Long Ju (MIT)")
print("=" * 60)
````
