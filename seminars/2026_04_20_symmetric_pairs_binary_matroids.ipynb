# Seminar Analysis: Symmetric Pairs & Binary Matroids

## Speaker's Claim

> *"A pair (g,k) of Lie algebras is a symmetric pair if k is the set of fixed points of an automorphism of g of order 2. We will show that when g and k are simply laced Lie algebras of equal rank, the pair gives rise to combinatorial structures whose numerical invariants can be predicted in terms of matroids that are representable over the field with two elements."*

**Joint work with:** Tianyuan Xu (University of Richmond)

## Background

| Concept | Definition |
|---------|------------|
| **Symmetric pair** | (g, k) where k = Fix(σ) for involution σ on g |
| **Simply laced** | ADE types — all roots same length |
| **Equal rank** | rank(g) = rank(k) |
| **Binary matroid** | Matroid representable over GF(2) |

````python
import numpy as np
import matplotlib.pyplot as plt

# Core functions
def angle_degrees(a, b):
    cos = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    return np.arccos(np.clip(cos, -1.0, 1.0)) * 180 / np.pi

def lock_status(angle):
    if angle < 10:
        return {"status": "VC / GOS", "color": "#2ecc71", "emoji": "🔒✅", "recommendation": "Accept"}
    elif angle < 35:
        return {"status": "weak VC", "color": "#f1c40f", "emoji": "⚠️", "recommendation": "Revise"}
    elif angle <= 55:
        return {"status": "IA / ZOS", "color": "#e74c3c", "emoji": "🔴💀", "recommendation": "Reject"}
    else:
        return {"status": "broken", "color": "#95a5a6", "emoji": "⚰️", "recommendation": "Discard"}

def hydration_score(angle):
    score = max(0, min(100, 100 * (1 - angle / 90)))
    status = lock_status(angle)
    return {
        'hydration_percent': round(score, 1),
        'status': status['status'],
        'recommendation': status['recommendation'],
        'emoji': status['emoji']
    }

print("=" * 60)
print("SEMINAR ANALYSIS: Symmetric Pairs & Binary Matroids")
print("=" * 60)
````

## Constraint Dimensions

| Dimension | Constraint | Score |

````python
# Claim vector (6 dimensions)
claim_vector = np.array([1.0, 1.0, 1.0, 0.9, 0.85, 1.0])
constraint_basis = np.ones(6)

angle = angle_degrees(claim_vector, constraint_basis)
status = lock_status(angle)
hydration = hydration_score(angle)

constraint_names = [
    "C1: Anchored to Helgason (symmetric pairs)",
    "C2: Anchored to ADE classification",
    "C3: Equal rank condition",
    "C4: Novelty (binary matroids connection)",
    "C5: Construction provided",
    "C6: Collaboration (Tianyuan Xu)"
]

print("\nConstraint adherence:\n")
for name, score in zip(constraint_names, claim_vector):
    bar = "█" * int(score * 20) + "░" * (20 - int(score * 20))
    print(f"  {name[:40]:40} {bar} {score:.0%}")

print(f"\n📐 Angle: {angle:.1f}°")
print(f"💧 Hydration: {hydration['hydration_percent']:.0f}%")
print(f"🏷️  Status: {status['status']} {status['emoji']}")
print(f"📋 Recommendation: {status['recommendation']}")
print(f"\n💡 This is a well-hydrated claim — classical anchoring, clear domain, novel connection.")
````

````python
fig, ax = plt.subplots(figsize=(10, 4))

angles_plot = np.linspace(0, 90, 100)
stability = np.cos(np.radians(angles_plot))

ax.plot(angles_plot, stability, 'k-', linewidth=2)
ax.axvline(angle, color=status['color'], linestyle='--', linewidth=2, label=f'Your claim: {angle:.1f}°')
ax.axvline(10, color='#2ecc71', linestyle=':', alpha=0.7, label='VC Lock (10°)')
ax.axvline(45, color='#e74c3c', linestyle='--', alpha=0.7, label='Critical (45°)')
ax.fill_between(angles_plot, 0, stability, where=(angles_plot < 10), alpha=0.2, color='#2ecc71')
ax.fill_between(angles_plot, 0, stability, where=(angles_plot > 35) & (angles_plot < 55), alpha=0.15, color='#e74c3c')

ax.set_xlim(0, 90)
ax.set_ylim(0, 1)
ax.set_xlabel('Angle from Constraint Basis (degrees)')
ax.set_ylabel('cos(θ) — stability')
ax.set_title('Hydration Analysis: Symmetric Pairs & Binary Matroids')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
