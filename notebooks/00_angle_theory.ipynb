## Core Hypothesis

> **A claim is consistent (VC/GOS) iff its representation is within 10° of its constraint basis.**  
> **At 45°, the claim is maximally unstable — it will collapse into IA/ZOS.**

This notebook visualizes the cosine constraint that underlies the `triplet-phase-lock` framework.

## The Cosine Constraint

The stability of a claim is given by $\cos(\theta)$, where $\theta$ is the angle between:
- The **claim vector** (what is being asserted)
- The **constraint basis** (the anchored structure it must respect)

$$\text{stability} = \cos(\theta) = \frac{\vec{c} \cdot \vec{b}}{|\vec{c}| |\vec{b}|}$$

```python
# Generate angles from 0° to 90°
angles = np.linspace(0, 90, 100)
stability = np.cos(np.radians(angles))  # Cosine constraint

# Create the visualization
plt.figure(figsize=(12, 5))

# Main curve
plt.plot(angles, stability, 'k-', linewidth=2.5, label='cos(θ)')

# Critical boundaries
plt.axvline(45, color='red', linestyle='--', linewidth=2, label='45° critical boundary (IA/ZOS)')
plt.axvline(10, color='green', linestyle=':', linewidth=2, label='10° VC lock zone')
plt.axvline(35, color='orange', linestyle=':', linewidth=1, alpha=0.5, label='Drift zone boundary')
plt.axvline(55, color='orange', linestyle=':', linewidth=1, alpha=0.5)

# Shade the 45° critical zone (±10°)
plt.fill_between(angles, 0, stability, where=(angles > 35) & (angles < 55), 
                  color='red', alpha=0.15, label='45° Critical Zone (Unstable)')

# Shade the VC lock zone
plt.fill_between(angles, 0, stability, where=angles < 10, 
                  color='green', alpha=0.15, label='VC Lock Zone (Stable)')

# Labels and title
plt.title("Cosine Constraint: Stability vs Angle", fontsize=14, fontweight='bold')
plt.xlabel("Angle between claim and constraint basis (degrees)", fontsize=12)
plt.ylabel("cos(θ) — alignment / stability", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper right')
plt.tight_layout()

# Annotate key points
plt.annotate('0°: Perfect lock\n(VC/GOS)', xy=(0, 1), xytext=(5, 0.85), 
             arrowprops=dict(arrowstyle='->', color='green'))
plt.annotate('45°: Maximum instability\n(IA/ZOS collapse)', xy=(45, 0.707), xytext=(50, 0.55),
             arrowprops=dict(arrowstyle='->', color='red'))
plt.annotate('90°: Orthogonal\n(No relationship)', xy=(90, 0), xytext=(75, 0.15),
             arrowprops=dict(arrowstyle='->', color='gray'))

plt.show()
```

## Interpretation Guide

| Angle Range | Status | Phase | Interpretation |
|-------------|--------|-------|----------------|
| 0° – 10° | **VC / GOS** | Locked | Valid construction — constraint preserved |
| 10° – 35° | Weak VC | Drifting | Partially constructed — risk of collapse |
| 35° – 55° | **IA / ZOS** | 45° critical zone | Invalid assignment — maximally unstable |
| 55° – 80° | Weak IA | Decoupled | Mostly unconstrained — noise dominant |
| 80° – 90° | Orthogonal | Irrelevant | No relationship — independent domain |

## Connection to Triplet-Phase-Lock

The progression through the triplet phases corresponds to angular drift:

- **Π⁽⁰⁾ expand** → Angle approaches 0° (anchored expansion)
- **Π⁽¹⁾ extend** → Angle drifts 10°–35° (extending without breaking lock)
- **Π⁽²⁾ resist** → Angle hits 45° critical zone (resistance reveals instability)
- **Π⁽³⁾ synthesis** → Returns to < 10° after correction (re-locking)

A claim that stays at 45° will follow the **ZOS path**: collapse → assignment → denial → inconsistency.
