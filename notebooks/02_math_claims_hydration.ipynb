3. Distinguish between **hydrated (VC)** and **dry (IA)** claims

## Core Hypothesis

> A mathematical claim is **hydrated (valid construction)** iff its angular distance from its constraint basis is **< 10°**.
> 
> A claim at **45°** represents an **invalid assignment** — stated without the constructive proof that would anchor it.

## 1. Setup and Core Functions

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from typing import Dict, List, Tuple

# Colab setup
if 'google.colab' in sys.modules:
    !pip install numpy matplotlib ipywidgets pandas > /dev/null 2>&1
    from IPython.display import display, HTML, Markdown
    print("✓ Colab environment ready")
else:
    print("✓ Local environment ready")

# Core functions (copied from Notebook 01 for self-containment)
def cosine_similarity(a: np.ndarray, b: np.ndarray) -> float:
    """Return cos(θ) between two vectors."""
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def angle_degrees(a: np.ndarray, b: np.ndarray) -> float:
    """Return angle in degrees between vectors a and b."""
    cos = cosine_similarity(a, b)
    return np.arccos(np.clip(cos, -1.0, 1.0)) * 180 / np.pi

def lock_status(angle: float) -> Dict[str, object]:
    """Classify angle into VC/IA/ZOS states."""
    if angle < 10:
        return {"status": "VC / GOS", "phase": "locked", "interpretation": "Valid construction — constraint preserved", "color": "green"}
    elif angle < 35:
        return {"status": "weak VC", "phase": "drifting", "interpretation": "Partially constructed — risk of collapse", "color": "yellowgreen"}
    elif angle <= 55:
        return {"status": "IA / ZOS", "phase": "45° critical zone", "interpretation": "Invalid assignment — unstable, prone to inconsistency", "color": "red"}
    elif angle < 80:
        return {"status": "weak IA", "phase": "decoupled", "interpretation": "Mostly unconstrained — noise dominant", "color": "orange"}
    else:
        return {"status": "orthogonal", "phase": "irrelevant", "interpretation": "No relationship — independent domain", "color": "gray"}

def claim_angle(claim_vector: np.ndarray, constraint_basis: np.ndarray) -> Dict[str, object]:
    """Given a claim and its supposed constraint basis, return the angle and lock status."""
    angle = angle_degrees(claim_vector, constraint_basis)
    status = lock_status(angle)
    status["angle_degrees"] = angle
    return status

print("✓ Core functions loaded")
```

## 2. Defining the Constraint Basis

For a mathematical claim about PCSP dichotomy to be **hydrated (VC)** , it must be anchored to:

| Dimension | Constraint | Description |
|-----------|------------|-------------|
| **C1** | Known CSP dichotomy theorem (Bulatov, Zhuk 2017) | The non-promise version is settled |
| **C2** | Specific finite algebra class (e.g., Mal'cev) | Not "all algebras" — a bounded domain |
| **C3** | Promise condition is explicitly defined | The gap between A and B is specified |
| **C4** | Construction exists (algorithm or reduction) | The proof is provided, not just asserted |
| **C5** | Acknowledges open status of general case | Honest boundary conditions |

We represent these as a 5-dimensional **constraint basis vector**.

```python
# Constraint basis for a hydrated PCSP claim
# Each dimension is binary: 1 = constraint respected, 0 = constraint absent

constraint_basis = np.array([
    1.0,  # C1: Anchored to known CSP dichotomy
    1.0,  # C2: Specific algebra class
    1.0,  # C3: Promise condition defined
    1.0,  # C4: Construction/proof provided
    1.0   # C5: Open general case acknowledged
])

print("Constraint basis vector:", constraint_basis)
print(f"Norm: {np.linalg.norm(constraint_basis):.2f}")
```

## 3. Test Case 1: The Bare PCSP Claim (Invalid Assignment)

**Claim:** *"The PCSP dichotomy is true for all finite algebras."*

This claim is **dry** — stated without proof, without anchoring to the known results, ignoring that the general case is open.

```python
# Represent the bare claim as a vector
# It asserts the conclusion (C4?) but lacks the constraints

bare_claim = np.array([
    0.0,  # C1: Ignores that CSP dichotomy is different from PCSP
    0.0,  # C2: Says "all finite algebras" — no specificity
    0.0,  # C3: Promise condition not engaged
    0.7,  # C4: Asserts truth but provides no construction (partial)
    0.0   # C5: Denies the open status by claiming it's solved
])

result = claim_angle(bare_claim, constraint_basis)

print("=" * 50)
print("TEST CASE 1: Bare PCSP Claim")
print("=" * 50)
print(f"Claim: 'The PCSP dichotomy is true for all finite algebras'")
print(f"\nAngle: {result['angle_degrees']:.1f}°")
print(f"Status: {result['status']}")
print(f"Phase: {result['phase']}")
print(f"Interpretation: {result['interpretation']}")
print(f"Verdict: INVALID ASSIGNMENT (IA/ZOS) — This claim will collapse under scrutiny")
```

## 4. Test Case 2: The Speaker's Actual Claim (Valid Construction)

**Claim:** *"I will present some dichotomy results for promise systems of equations over various classes of algebras, including Mal'cev algebras. The general question remains open."*

This claim is **hydrated** — it respects all constraints and acknowledges boundaries.

```python
# Represent the speaker's actual claim

speakers_claim = np.array([
    1.0,  # C1: Builds on known CSP dichotomy
    0.8,  # C2: Specific classes (Mal'cev et al.) — not all
    0.9,  # C3: Promise condition is central to the talk
    0.9,  # C4: Will present actual proofs (algorithm/reduction)
    1.0   # C5: Explicitly says "open in general"
])

result = claim_angle(speakers_claim, constraint_basis)

print("=" * 50)
print("TEST CASE 2: Speaker's Actual Claim")
print("=" * 50)
print(f"Claim: 'Some dichotomy results for Mal'cev algebras and others; general case open'")
print(f"\nAngle: {result['angle_degrees']:.1f}°")
print(f"Status: {result['status']}")
print(f"Phase: {result['phase']}")
print(f"Interpretation: {result['interpretation']}")
print(f"Verdict: VALID CONSTRUCTION (VC/GOS) — This claim is hydrated and stable")
```

## 5. Test Case 3: Over-Extrapolation (Danger Zone)

**Claim:** *"Since Mal'cev algebras have a dichotomy, all algebras do."*

This is a logical leap — extending results beyond their constraint boundaries.

```python
# Over-extrapolation claim

over_extrapolation = np.array([
    0.6,  # C1: Partial anchor — knows CSP dichotomy exists
    0.2,  # C2: Extends from specific to "all" — weak
    0.3,  # C3: Promise condition gets lost
    0.4,  # C4: No construction for the generalization
    0.1   # C5: Ignores open status
])

result = claim_angle(over_extrapolation, constraint_basis)

print("=" * 50)
print("TEST CASE 3: Over-Extrapolation")
print("=" * 50)
print(f"Claim: 'Since Mal'cev algebras have a dichotomy, all algebras do'")
print(f"\nAngle: {result['angle_degrees']:.1f}°")
print(f"Status: {result['status']}")
print(f"Phase: {result['phase']}")
print(f"Interpretation: {result['interpretation']}")
print(f"Verdict: CRITICAL ZONE — This claim is at 45° and will collapse")
```

## 6. Visualization: All Claims on the Angle Spectrum

```python
# Collect all results
claims = [
    ("Bare PCSP claim (IA)", bare_claim),
    ("Speaker's actual claim (VC)", speakers_claim),
    ("Over-extrapolation (45° critical)", over_extrapolation)
]

angles = []
names = []
colors = []

for name, vec in claims:
    ang = angle_degrees(vec, constraint_basis)
    angles.append(ang)
    names.append(name)
    colors.append(lock_status(ang)['color'])

# Create visualization
fig, ax = plt.subplots(figsize=(12, 6))

# Background zones
ax.axvspan(0, 10, alpha=0.2, color='green', label='VC Lock Zone (0°-10°)')
ax.axvspan(35, 55, alpha=0.2, color='red', label='45° Critical Zone (35°-55°)')
ax.axvspan(80, 90, alpha=0.1, color='gray', label='Orthogonal (>80°)')

# Plot each claim as a vertical line
for i, (name, ang, color) in enumerate(zip(names, angles, colors)):
    y_pos = 0.8 - i * 0.25
    ax.scatter(ang, y_pos, s=200, c=color, marker='D', edgecolors='black', zorder=5)
    ax.annotate(name, xy=(ang, y_pos), xytext=(5, 0), textcoords='offset points', 
                fontsize=10, ha='left', va='center')

# Critical boundaries
ax.axvline(10, color='green', linestyle='--', alpha=0.7, linewidth=1.5)
ax.axvline(35, color='orange', linestyle=':', alpha=0.7)
ax.axvline(45, color='red', linestyle='--', alpha=0.7, linewidth=2, label='45° Boundary')
ax.axvline(55, color='orange', linestyle=':', alpha=0.7)

ax.set_xlim(0, 90)
ax.set_ylim(0, 1)
ax.set_xlabel("Angle from Constraint Basis (degrees)", fontsize=12)
ax.set_title("Math Claims: Hydration Status by Angle", fontsize=14, fontweight='bold')
ax.legend(loc='upper right')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print summary table
print("\n" + "=" * 60)
print("SUMMARY: Hydration Test Results")
print("=" * 60)
for name, ang, color in zip(names, angles, colors):
    status = lock_status(ang)['status']
    print(f"{name:35} → {ang:5.1f}°  [{status}]")
```

## 7. Interactive Demo: Test Your Own Claim

Use the widget below to create a claim vector and see where it lands.

```python
from ipywidgets import interact, FloatSlider, VBox, Label, HBox

def test_claim(c1=0.5, c2=0.5, c3=0.5, c4=0.5, c5=0.5):
    claim_vec = np.array([c1, c2, c3, c4, c5])
    angle = angle_degrees(claim_vec, constraint_basis)
    status = lock_status(angle)
    
    print(f"\nAngle: {angle:.1f}°")
    print(f"Status: {status['status']}")
    print(f"Phase: {status['phase']}")
    print(f"Interpretation: {status['interpretation']}")
    
    # Simple bar chart
    fig, ax = plt.subplots(figsize=(8, 3))
    colors_bar = ['green' if x > 0.7 else 'red' if x < 0.3 else 'orange' for x in [c1, c2, c3, c4, c5]]
    ax.bar(['C1\n(Anchored)', 'C2\n(Specific)', 'C3\n(Promise)', 'C4\n(Proof)', 'C5\n(Open)'], 
           [c1, c2, c3, c4, c5], color=colors_bar, alpha=0.7)
    ax.set_ylim(0, 1)
    ax.set_ylabel("Constraint Respect")
    ax.set_title(f"Claim Profile — {status['status']}")
    plt.show()

print("Adjust the sliders to create your own mathematical claim:")
print("1.0 = fully respects the constraint, 0.0 = ignores it completely\n")

interact(test_claim,
         c1=FloatSlider(min=0, max=1, step=0.1, value=0.5, description='C1: Anchored'),
         c2=FloatSlider(min=0, max=1, step=0.1, value=0.5, description='C2: Specific'),
         c3=FloatSlider(min=0, max=1, step=0.1, value=0.5, description='C3: Promise'),
         c4=FloatSlider(min=0, max=1, step=0.1, value=0.5, description='C4: Proof'),
         c5=FloatSlider(min=0, max=1, step=0.1, value=0.5, description='C5: Open status'))
```

## Connection to the Seminar

The speaker's actual position — *"Some dichotomy results for Mal'cev algebras; general case open"* — is **hydrated (VC)** because:

1. It anchors to known CSP dichotomy (C1 ✓)
2. It specifies the algebra class (C2 ✓)
3. The promise condition is central (C3 ✓)
4. The talk will present actual constructions (C4 ✓)
5. It explicitly acknowledges the open general case (C5 ✓)

The **invalid assignment** would be claiming the general dichotomy is solved — a 45° claim that collapses under the cosine constraint.
