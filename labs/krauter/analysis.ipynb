## The Scientific Arc: From "Nurture Trumps Nature" to Detecting Heritability

**Early finding (2012):** A twin study published in *Genome Research* found that salivary microbes of identical twins were **not significantly more similar** to each other than those of fraternal twins. The conclusion: "The human genome does not significantly affect which bacteria are living in a person's mouth. It appears to be more of an environmental effect." [citation:2][citation:10]

**More nuanced understanding:** Later research using larger sample sizes revealed that **heritability is actually significant** — but requires larger studies to detect. As the lab's research page states: "MZs (identical twins) are more similar than other pairwise comparisons. This can ONLY be explained by heritability — i.e. human genes determine this." [citation:4]

**The explanation:** The initial 2012 study was underpowered to detect heritable effects. A follow-up GWAS on ~750 individuals identified "suggestively significant" peaks, but the lab estimates that increasing the study population by **12X** will enable statistically significant results. [citation:4]

````python
import numpy as np
import matplotlib.pyplot as plt

# Core functions
def angle_degrees(a, b):
    cos = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    return np.arccos(np.clip(cos, -1.0, 1.0)) * 180 / np.pi

def lock_status(angle):
    if angle < 10:
        return {"status": "VC / GOS", "color": "#2ecc71", "emoji": "🔒✅", "recommendation": "Well-constructed research program"}
    elif angle < 35:
        return {"status": "weak VC", "color": "#f1c40f", "emoji": "⚠️", "recommendation": "Needs clearer methods/metrics"}
    elif angle <= 55:
        return {"status": "IA / ZOS", "color": "#e74c3c", "emoji": "🔴💀", "recommendation": "Vague claims, likely over-hyped"}
    else:
        return {"status": "broken", "color": "#95a5a6", "emoji": "⚰️", "recommendation": "No clear research direction"}

def hydration_score(angle):
    score = max(0, min(100, 100 * (1 - angle / 90)))
    status = lock_status(angle)
    return {
        'hydration_percent': round(score, 1),
        'status': status['status'],
        'recommendation': status['recommendation'],
        'emoji': status['emoji']
    }

print("=" * 60)
print("LAB ANALYSIS: Krauter Lab")
print("PI: Kenneth S. Krauter (Professor Emeritus, MCDB)")
print("=" * 60)
````

````python
# Claim vector (8 dimensions)
claim_vector = np.array([1.0, 1.0, 1.0, 0.9, 1.0, 0.8, 0.85, 1.0])
constraint_basis = np.ones(8)

angle = angle_degrees(claim_vector, constraint_basis)
status = lock_status(angle)
hydration = hydration_score(angle)

constraint_names = [
    "D1: Problem clearly stated",
    "D2: Specific biological focus",
    "D3: Methods specified",
    "D4: Key findings",
    "D5: Interdisciplinary",
    "D6: PI status (Emeritus)",
    "D7: Publication record",
    "D8: Real-world relevance"
]

print("\nConstraint adherence:\n")
for name, score in zip(constraint_names, claim_vector):
    bar = "█" * int(score * 20) + "░" * (20 - int(score * 20))
    print(f"  {name[:25]:25} {bar} {score:.0%}")

print(f"\n📐 Angle: {angle:.1f}°")
print(f"💧 Hydration: {hydration['hydration_percent']:.0f}%")
print(f"🏷️  Status: {status['status']} {status['emoji']}")
print(f"📋 Recommendation: {status['recommendation']}")
print(f"\n💡 Well-hydrated research program — clear problem focus, rigorous twin study methods, and pioneering work on microbiome genetics. The Professor Emeritus status and less recent publication activity slightly lower the hydration compared to active junior faculty labs.")
````

````python
fig, ax = plt.subplots(figsize=(10, 4))

angles_plot = np.linspace(0, 90, 100)
stability = np.cos(np.radians(angles_plot))

ax.plot(angles_plot, stability, 'k-', linewidth=2)
ax.axvline(angle, color=status['color'], linestyle='--', linewidth=2, label=f'Krauter Lab: {angle:.1f}°')
ax.axvline(10, color='#2ecc71', linestyle=':', alpha=0.7, label='VC Lock (10°)')
ax.axvline(45, color='#e74c3c', linestyle='--', alpha=0.7, label='Critical (45°)')
ax.fill_between(angles_plot, 0, stability, where=(angles_plot < 10), alpha=0.2, color='#2ecc71')
ax.fill_between(angles_plot, 0, stability, where=(angles_plot > 35) & (angles_plot < 55), alpha=0.15, color='#e74c3c')

ax.set_xlim(0, 90)
ax.set_ylim(0, 1)
ax.set_xlabel('Angle from Constraint Basis (degrees)')
ax.set_ylabel('cos(θ) — stability')
ax.set_title('Hydration Analysis: Krauter Lab')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````
